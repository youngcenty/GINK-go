```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
#  DISC1
gene_chrom, gene_start, gene_end = 'chr1', 231626790, 232041272

# PAX6
gene_chrom, gene_start, gene_end = 'chr11', 31784779, 31817961

# SOX2
gene_chrom, gene_start, gene_end = 'chr3', 181711925, 181714436

# YWHAG
gene_chrom, gene_start, gene_end = 'chr7', 76326799, 76358991

# HMGB1
gene_chrom, gene_start, gene_end = 'chr13', 30456704, 30617597


# up and down 1M
region_start = gene_start - 1_000_000
region_end = gene_end + 1_000_000
gene_chrom, gene_start, gene_end, region_start, region_end
```

```python
def calculate_min_distance(row, gene_start, gene_end):
    distances = [
        abs(row['Start'] - gene_start),
        abs(row['End'] - gene_start),
        abs(row['Start'] - gene_end),
        abs(row['End'] - gene_end)
    ]
    return min(distances)


res = []
for histone in ["ATAC", "H3K27ac", "H3K4me1", "H3K27me3","H3K4me3",  "H3K36me3", "H3K9me3"]: #,   
    df = pd.read_csv(f"/storage/zhangyanxiaoLab/qihongjian/github/zhanglab-code/projects/transcriptome_4D/code/ESC/diff_enhancers/{histone}_bcv0.4_diff_regions.csv")
    df = df[(df["Chr"]==gene_chrom) & (df["Start"] >= region_start) & (df["End"] <= region_end)]
    df['distance'] = df.apply(calculate_min_distance, axis=1, gene_start=gene_start, gene_end=gene_end)

    df["histone"] = histone
    df.columns = ['Geneid', 'logFC', 'logCPM', 'PValue', 'FDR', 'D0-cpm', 'D3-cpm', 'Chr',
       'Start', 'End', 'Strand', 'Length', 'ESC_D0-counts', 'ESC_D3-counts', 'distance', 'histone']
    res.append(df)
```

```python
df = pd.concat(res).sort_values("distance")
```

```python
cols = ['histone', 'logFC', 'PValue', 'FDR', 'D0-cpm', 'D3-cpm', 'Chr',
       'Start', 'End',  'Length', 'distance', ]

df[df["histone"]!="H3K9me3"][cols]
df[cols].sort_values("Start")
```
